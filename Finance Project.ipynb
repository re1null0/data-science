```python
import pandas as pd
from pandas_datareader import data, wb
import numpy as np
import seaborn as sns
from datetime import datetime
import os
os.environ["IEX_API_KEY"] = "pk_xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx"
%matplotlib inline
```

```python
bac = data.DataReader("BAC", 'yahoo', datetime(2006, 1, 1), datetime(2016, 1, 1))
citi = data.DataReader("C", 'yahoo', datetime(2006, 1, 1), datetime(2016, 1, 1))
goldman = data.DataReader("GS", 'yahoo', datetime(2006, 1, 1), datetime(2016, 1, 1))
chase = data.DataReader("JPM", 'yahoo', datetime(2006, 1, 1), datetime(2016, 1, 1))
stanley = data.DataReader("MS", 'yahoo', datetime(2006, 1, 1), datetime(2016, 1, 1))
fargo = data.DataReader("WFC", 'yahoo', datetime(2006, 1, 1), datetime(2016, 1, 1))
```

```python
tickers = sorted(['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC'])
```

```python
bank_stocks = pd.concat([bac, citi, goldman, chase, stanley, fargo], keys=tickers, axis=1)
```


$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
returns = pd.DataFrame()
```

```python
returns = bank_stocks.xs('Close', axis=1, level=1).pct_change()
for bank in returns.columns:
    returns[bank.split(' ')[0]+' Return'] = returns[bank]
    returns.drop(bank, axis=1, inplace=True)
returns
```
